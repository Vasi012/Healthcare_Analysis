# Section 2 - Transformation

During the transformation stage, the healthcare dataset will be cleaned and prepared for analysis. This included handling missing values, removing duplicate records, correcting inconsistent data formats, and standardising column names and data types. Where appropriate, additional derived fields were created to support analysis, ensuring the dataset was accurate, consistent, and suitable for reporting.

```python
#Missing Values
print("Missing Values:")
missing = df_raw.isnull().sum()
missing_pct = (missing / len(df_raw) * 100).round(2)
audit = pd.DataFrame({"Missing Values": missing, "Percentage": missing_pct})

if audit["Missing Values"].sum() > 0:
    print(audit[audit["Missing Values"] > 0])
else:
    print("No missing values found in the dataset.")

#Duplicated Rows
dupes = df_raw.duplicated().sum()
print(f"Duplicate Records: {dupes}")

#Unique values per column
print("Unique Values:")
print(df_raw.nunique())
```

```python
# Categorical Value Distribution
for col in ["sex", "smoker", "region"]:
    print(f"[{col}] Value Distribution:")
    print(df_raw[col].value_counts())
    print()
```

```python
# Work on a copy, never mutate raw data.

df = df_raw.copy()

#Standardise string column
str_cols = ["sex", "smoker", "region"]
for col in str_cols:
    df[col] = df[col].str.lower().str.strip()

print(" String Columns are now Standardised.")

# Remove duplicates if any
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)
print(f"Records before duplicate removal: {before}")
print(f"Records after duplicate removal: {after}")

# Validate Numeric Ranges
age_range = df[(df["age"] <18) | (df["age"] > 110)]
bmi_range = df[(df["bmi"] < 10) | (df["bmi"] > 60)]
charges_range = df[df["charges"] <=0]

print(f"Records with age outside valid range: {len(age_range)}")
print(f"Records with BMI outside valid range: {len(bmi_range)}")
print(f"Records with charges outside valid range: {len(charges_range)}")
```

## Feature Engineering

```python
# Binary encode: female = 0, male = 1
df["sex_encoded"] = df["sex"].map({"female": 0, "male": 1})

print(df[["sex", "sex_encoded"]].drop_duplicates().sort_values("sex_encoded"))
```

```python
#One-hot encode: region
region_dummies = pd.get_dummies(df["region"], prefix="region", drop_first=True)
df = pd.concat([df, region_dummies], axis=1)

print(df[["region"] + list(region_dummies.columns)].drop_duplicates().sort_values("region"))
```

```python
#binary encode: no = 0, yes = 1
df["smoker_encoded"] = df["smoker"].map({"no": 0, "yes": 1})
print(df[["smoker", "smoker_encoded"]].drop_duplicates().sort_values("smoker_encoded"))
```

```python
#BMI Category - Standard Classification
def bmi_category(bmi):
    if bmi < 18.5:
        return "underweight"
    elif 18.5 <= bmi < 25:
        return "normal"
    elif 25 <= bmi < 30:
        return "overweight"
    else:
        return "obese"


df["bmi_category"] = df["bmi"].apply(bmi_category)
df["bmi_cat_encoded"] = df["bmi_category"].map({"underweight": 0, "normal": 1, "overweight": 2, "obese": 3})

print(df["bmi_category"].value_counts())
```

```python
#Age Groups
age_bins = [0, 25, 35, 50, 65, 110]
age_labels = ["18-24", "25-34", "35-49", "50-64", "65+"]
df["age_group"] = pd.cut(df["age"], bins=age_bins, labels=age_labels, right=False)

print(df["age_group"].value_counts().sort_index())
```

```python
#Smoker + Obese high-risk group
df["smoker_obese"] = (df["smoker_encoded"] == 1) & (df["bmi"] >= 30).astype(int)

print(f"High-risk group (smoker + obese): {df['smoker_obese'].sum()} records")
print(f"As % of total records: {(df['smoker_obese'].sum() / len(df) * 100):.2f}%")
```

```python
#Log Transform charges
df["log_charges"] = np.log1p(df["charges"])  # Use log1p to handle zero charges if any
print(f" Charges skewness before log transform: {df['charges'].skew():.3f}")
print(f"Charges skewness after log transform: {df['log_charges'].skew():.3f}")

# Quick visual coparaison
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

sns.histplot(df["charges"], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title("Distribution of Charges (Before Log Transform)")
axes[0].set_xlabel("Charges")

sns.histplot(df["log_charges"], bins=30, kde=True, ax=axes[1], color='salmon')
axes[1].set_title("Distribution of Charges (After Log Transform)")
axes[1].set_xlabel("Log(Charges + 1)")

plt.suptitle("Distribution before and after log transform")
plt.tight_layout()
plt.show()
```
